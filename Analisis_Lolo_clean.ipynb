```python
#Armo funcion para graficar en el plano

def graficar_plano(coordenadas, anio):

    fondo = plt.figure()
    #fondo.patch.set_facecolor('xkcd:light grey')
    plt.title('Distribucion geogrfica de anuncios publicados en '+anio)
    
    plt.grid(color='black', linestyle='-', linewidth=1, alpha = 0.2, zorder = 0)
    
    #Asigno en ejes
    y = coordenadas['lat']
    x = coordenadas['lng']

    #Ploteo en rojo
    plt.scatter(x, y, color = "r", alpha = 0.30, s = 15,zorder = 2)


    #Importo coordenadas de frontera de mexico

    #Asigno ejes
    y2 = mex['lat']
    x2 = mex['lng']

    #Ploteo en negro
    plt.scatter(x2, y2, color = 'black', alpha = 0.5, s=0.3, zorder = 4)
    
    plt.xlabel("Latitud")
    plt.ylabel("Longitud")
    
    plt.show()
    fondo.savefig('distribucion'+anio+'.png')
```

```python
# Voy a estudiar el % de completitud de las categorias de interes ao a ao
df['fecha'] = pd.to_datetime(df['fecha'])

nulos_por_ao = df.loc[:,['ao', 'titulo',	'direccion',	'antiguedad',	'habitaciones',	'garages',	'banos'
                          ,'metrostotales',	'lat']]


nulos_por_ao['ao'] = df['fecha'].dt.year
nulos_por_ao.head()

def contar_nulos(serie):
    
    serie = serie.apply(lambda x: 1 if pd.isnull(x) else 0)#.apply(hacer_dos)
    return np.sum(serie)
    
nulos_por_ao_cat = nulos_por_ao.groupby('ao').agg(contar_nulos).reset_index()
totales_por_ao = nulos_por_ao.groupby('ao')['lat'].agg(len).reset_index()
totales_por_ao = totales_por_ao.rename(columns={'lat' : 'totales'})

pct_nulos_por_ao_cat = pd.merge(nulos_por_ao_cat, totales_por_ao, on = 'ao', how = 'inner')

def sacar_porcentaje(serie):
    
    copia_aos = serie['ao']
    copia_totales = serie['totales']
    serie_porcentual = serie / copia_totales * 100
    
    serie_porcentual['ao'] = copia_aos
    serie_porcentual['totales'] = copia_totales
    
    return serie_porcentual

pct_nulos_por_ao_cat = pct_nulos_por_ao_cat.apply(sacar_porcentaje, axis = 1)
pct_nulos_por_ao_cat['ao'] = pct_nulos_por_ao_cat['ao'].astype(np.int32)
pct_nulos_por_ao_cat['totales'] = pct_nulos_por_ao_cat['totales'].astype(np.int32)
```
